# Univariate model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(column, target, dataframe):
    np.random.seed(0)
    shuffled_index = np.random.permutation(dataframe.index)
    random_dataframe = dataframe.reindex(shuffled_index)
    
    last_train_row = int(len(random_dataframe)/2)
    train_set = random_dataframe.iloc[0:last_train_row]
    test_set = random_dataframe.iloc[last_train_row:]
    
    knn = KNeighborsRegressor()
    knn.fit(train_set[[column]], train_set[target])
    
    predicted_labels = knn.predict(test_set[[column]])
    mse = mean_squared_error(test_set[target], predicted_labels)
    rmse = np.sqrt(mse)
    
    return rmse
```

```python
knn_train_test('length', 'price', numeric_cars)
```

```python
rmse_results = {}
column_considered = list()
for column in numeric_cars.columns:
    column_considered.append(column)
column_considered.remove('price')
```

```python
for column in column_considered:
    rmse_results[column] = knn_train_test(column, 'price', numeric_cars)
```

```python
rmse_results
```

```python
rmse_series = pd.Series(rmse_results)
rmse_series = rmse_series.sort_values()
rmse_series
```

```python
def knn_train_test2(column, target, dataframe):
    np.random.seed(0)
    shuffled_index = np.random.permutation(dataframe.index)
    random_dataframe = dataframe.reindex(shuffled_index)
    
    last_train_row = int(len(random_dataframe)/2)
    train_set = random_dataframe.iloc[0:last_train_row]
    test_set = random_dataframe.iloc[last_train_row:]
    
    k_values = [1, 3, 5, 7, 9]
    k_rmse = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_set[[column]], train_set[target])

        predicted_labels = knn.predict(test_set[[column]])
        mse = mean_squared_error(test_set[target], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmse[k] = rmse
    return k_rmse
```

```python
feature_k_rmse = {}
for column in column_considered:
    rmse_results = knn_train_test2(column, 'price', numeric_cars)
    feature_k_rmse[column] = rmse_results
feature_k_rmse
```

```python
import matplotlib.pyplot as plt

best_features = {}
plt.figure(figsize=(10,12))

for key, value in feature_k_rmse.items():
    x = list(value.keys())
    y = list(value.values())
    
    order = np.argsort(x)
    x_ordered = np.array(x)[order]
    y_ordered = np.array(y)[order]
    print(key)
    print('average rmse: '+str(np.mean(y)))
    best_features[key] = np.mean(y)
    
    plt.plot(x_ordered, y_ordered, label=key)
    plt.xlabel('K value')
    plt.ylabel('RMSE')
plt.legend()
plt.show()
```

```python
sorted_features_list = sorted(best_features, key=best_features.get)
sorted_features_list
```

```python
def knn_train_test3(columns, target, dataframe):
    np.random.seed(0)
    shuffled_index = np.random.permutation(dataframe.index)
    random_dataframe = dataframe.reindex(shuffled_index)
    
    last_train_row = int(len(random_dataframe)/2)
    train_set = random_dataframe.iloc[0:last_train_row]
    test_set = random_dataframe.iloc[last_train_row:]

    k_rmse = {}
    
    for k in range(1,26):
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_set[columns], train_set[target])

        predicted_labels = knn.predict(test_set[columns])
        mse = mean_squared_error(test_set[target], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmse[k] = rmse
    return k_rmse
```
