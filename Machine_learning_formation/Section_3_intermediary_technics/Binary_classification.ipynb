```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
```

```python
admissions = pd.read_csv('admissions.csv')
```

```python
admissions.info()
```

```python
plt.scatter(x=admissions['gpa'], y=admissions['admit'])
plt.xlabel('GPA')
plt.ylabel('Admit')
plt.show()
```

# Logistic function

```python
def logistic(x):
    return np.exp(x)/(1 + np.exp(x))
```

```python
x = np.linspace(-6, 6, 50, dtype=float)
y = logistic(x)
plt.plot(x, y)
plt.ylabel('Probabilities')
plt.show()
```

# Logistic regression model

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(solver='liblinear')
logistic_model.fit(admissions[['gpa']], admissions['admit'])
```

# Probabilities calcul

```python
pred_prod = logistic_model.predict_proba(admissions[['gpa']])
```

```python
plt.scatter(x=admissions['gpa'], y=pred_prod[:,1])
plt.xlabel('GPA')
plt.ylabel('Predicted probabilities of being accepted')
plt.show()
```

# Predictions

```python
fitted_labels = logistic_model.predict(admissions[['gpa']])
```
