# Setup K Means algorithm

```python
num_cluster = 5
```

```python
random_initial_points = np.random.choice(point_guards.index, size = num_cluster)
```

```python
random_initial_points
```

```python
centroids = point_guards.loc[random_initial_points]
```

```python
centroids
```

```python
plt.scatter(point_guards['ppg'], point_guards['atr'], c='yellow')
plt.scatter(centroids['ppg'], centroids['atr'], c='red')
plt.title('Points guards')
plt.xlabel('Points per game')
plt.ylabel('Assists Turn over ratio')
plt.show()
```

```python
def centroids_to_dict(centroids):
    dictionary = dict()
    counter = 0
    for index, row in centroids.iterrows():
        coordinates = [row['ppg'], row['atr']]
        dictionary[counter] = coordinates
        counter += 1
    return dictionary
```

```python
centroids_dict = centroids_to_dict(centroids)
```

```python
centroids_dict
```

# Euclid distance

```python
import math
```

```python
def calculate_distance(centroid, player_value):
    root_distance = 0
    for x in range(0, len(centroid)):
        difference = centroid[x] - player_value[x]
        squared_difference = difference ** 2
        root_distance += squared_difference
    euclid_distance = math.sqrt(root_distance)
    return euclid_distance
```

```python
def assign_to_cluster(row):
    lowest_distance = -1
    closest_cluster = -1
    
    for cluster_id, centroid in centroids_dict.items():
        df_row = [row['ppg'], row['atr']]
        euclidian_distance = calculate_distance(centroid, df_row)
        
        if lowest_distance == -1:
            lowest_distance = euclidian_distance
            closest_cluster = cluster_id
        elif euclidian_distance < lowest_distance:
            lowest_distance = euclidian_distance
            closest_cluster = cluster_id
    return closest_cluster
```

```python
point_guards['cluster'] = point_guards.apply(lambda row: assign_to_cluster(row), axis=1)
```

```python
point_guards
```

```python
def visualize_cluster(df, num_cluster):
    colors = ['b', 'g', 'r', 'c', 'm', 'y' ,'k']
    for n in range(num_cluster):
        clustered_df = df[df['cluster'] == n]
        plt.scatter(clustered_df['ppg'], clustered_df['atr'], c=colors[n-1])
        plt.xlabel('Points per match')
        plt.ylabel('Assists Turn over ratio')
    plt.show()
```

```python
visualize_cluster(point_guards, 5)
```

```python
def recalculate_centroid(df):
    new_centroid_dict = dict()
    for cluster_id in range(0, num_cluster):
        values_in_cluster = df[df['cluster'] == cluster_id]
        new_centroid = [np.average(values_in_cluster['ppg']), np.average(values_in_cluster['atr'])]
        new_centroid_dict[cluster_id] = new_centroid
    return new_centroid_dict
```

```python
centroids_dict = recalculate_centroid(point_guards)
```

```python
centroids_dict
```

```python
point_guards['cluster'] = point_guards.apply(lambda row: assign_to_cluster(row), axis=1)
```

```python
visualize_cluster(point_guards, num_cluster)
```

```python
centroids_dict = recalculate_centroid(point_guards)
point_guards['cluster'] = point_guards.apply(lambda row: assign_to_cluster(row), axis=1)
visualize_cluster(point_guards, num_cluster)
```

# Using K Means class from sklearn

```python
point_guards = nba[nba['pos'] == 'PG']
point_guards['ppg'] = point_guards['pts']/point_guards['g']
point_guards = point_guards[point_guards['tov'] != 0]
point_guards['atr'] = point_guards['ast']/point_guards['tov']
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=num_cluster)
kmeans.fit(point_guards[['ppg', 'atr']])
point_guards['cluster'] = kmeans.labels_

visualize_cluster(point_guards, num_cluster)
```
