```python
import matplotlib.pyplot as plt 
import math
x = [0.1*i for i in range(64)] 
square = [(i**2) for i in x]
exponential = [math.exp(i) for i in x]
plt.plot(x, square, label='x^2', color='blue', linewidth=2.5)
plt.plot(x, exponential, label='exp(x)', color='red', linewidth=0.5)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```
